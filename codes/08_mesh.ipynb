```python
%load_ext autoreload
%autoreload 2
%matplotlib inline 

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import healpy as hp
from scipy.spatial import ConvexHull
import meshio
```

```python
nside = 1
indexes = range(nside**2 * 12)
x, y, z = hp.pix2vec(nside, indexes)
coords = np.vstack([x, y, z]).transpose()
coords = np.asarray(coords)

tri = ConvexHull(coords)
tri.simplices.shape

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(221, projection='3d')
for simplex in tri.simplices:
    s = simplex.copy()
    s = np.append(s, simplex[0])
    ax.plot(coords[s][:,0], coords[s][:,1], coords[s][:,2], 'b--', LineWidth=0.3)

nside = 2
indexes = range(nside**2 * 12)
x, y, z = hp.pix2vec(nside, indexes)
coords = np.vstack([x, y, z]).transpose()
coords = np.asarray(coords)

tri = ConvexHull(coords)
tri.simplices.shape

ax = fig.add_subplot(222, projection='3d')
for simplex in tri.simplices:
    s = simplex.copy()
    s = np.append(s, simplex[0])
    ax.plot(coords[s][:,0], coords[s][:,1], coords[s][:,2], 'b--', LineWidth=0.3)

nside = 4
indexes = range(nside**2 * 12)
x, y, z = hp.pix2vec(nside, indexes)
coords = np.vstack([x, y, z]).transpose()
coords = np.asarray(coords)
print(coords)
tri = ConvexHull(coords)
tri.simplices.shape

ax = fig.add_subplot(223, projection='3d')
for simplex in tri.simplices:
    s = simplex.copy()
    s = np.append(s, simplex[0])
    ax.plot(coords[s][:,0], coords[s][:,1], coords[s][:,2], 'b--', LineWidth=0.3)

nside = 8
indexes = range(nside**2 * 12)
x, y, z = hp.pix2vec(nside, indexes)
coords = np.vstack([x, y, z]).transpose()
coords = np.asarray(coords)

tri = ConvexHull(coords)
tri.simplices.shape

ax = fig.add_subplot(224, projection='3d')
for simplex in tri.simplices:
    s = simplex.copy()
    s = np.append(s, simplex[0])
    ax.plot(coords[s][:,0], coords[s][:,1], coords[s][:,2], 'b--', LineWidth=0.3)
```

```python
for i in range(10):
    nside = 2**i
    indexes = range(nside**2 * 12)
    x, y, z = hp.pix2vec(nside, indexes)
    coords = np.vstack([x, y, z]).transpose()
    coords = np.asarray(coords)

    tri = ConvexHull(coords)
    tri.simplices.shape

    cells = {"triangle": tri.simplices}
    meshio.write_points_cells(
        "09_meshes/HEALPix_{}.xml".format(nside),
        coords,
        cells,
        )
```
